# Personalality Types Prediction and Analyzation 

# Import Libaries 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

# Load Data 

```python
data = pd.read_csv('data.csv')

data.head()
```

# Explore Data

```python
data.info()
data.describe()
data.isnull().sum()
```

# Data Cleaning 

```python
# Example: Drop rows with missing values (if any)
data = data.dropna()
# Example: Check for duplicates
data = data.drop_duplicates()
```

# Data Visualization

```python
# age distributions and counts
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
# count of different personality types
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Personality', order=data['Personality'].value_counts().index)
plt.title('Count of Personality Types')
plt.show()
# relationships between scores and personality types 
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Personality', y='Introversion Score')
plt.title('Introversion Scores by Personality Type')
plt.show()
# heatmap to visualize correlations between numerical variables 
score_columns = ['Introversion Score', 'Sensing Score', 'Thinking Score', 'Judging Score']
average_scores = data.groupby('Personality')[score_columns].mean() # getting the average score based on personality type
# UNCOMMENT TO SWITCH X AND Y AXIS ------ average_scores_transposed = average_scores.transpose()
plt.figure(figsize=(10, 6))
sns.heatmap(average_scores, annot=True, cmap='coolwarm', linewidths=0.5) 
plt.title('Average Personality Trait Scores by Personality Type') 
plt.xlabel('Trait')
plt.ylabel('Personality Type')
plt.show()
# Bar plot to show different types of personalities average educational level
average_education = data.groupby('Personality')['Education'].mean() # get average educational level by personality 
plt.figure(figsize=(10, 6))
sns.barplot(x='Personality', y='Education', data=average_education))
plt.title('Average Educational Level based on Personality Type')
plt.xlabel('Education Level')
plt.ylabel('Personality Type')
plt.show()
```
